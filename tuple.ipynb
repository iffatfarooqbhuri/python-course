```python
students = [("Ali", 85), ("Sara", 77), ("Iffat", 100), ("Faiza", 90)]
students.sort(key=lambda x: x[1])
print("By marks ascendimg:", students)
students.sort(key=lambda x: x[1],reverse=True)
print("By marks descending:", students)
students.sort(key=lambda x: x[0])
print("By names:", students)
```
